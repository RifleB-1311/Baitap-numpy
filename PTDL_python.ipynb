Chạy dữ liệu.

```python
import pandas as pd
import numpy as np
data= pd.read_excel("/content/Dataset-Temperature-by-years.xls")
print(data)
```

Phân tích max, min, median, phương sai, hiệu chỉnh, tứ phân vị.

```python
import statistics
# Cột độ mặn (Salinity)
print("Max Salinity: ",max(data.iloc[:,3]))  #Max
print("Min Salinity: ",min(data.iloc[:,3]))  #Min
print("Median Salinity: ",data['Salinity'].median())  #Median 
print("Phương sai chưa hiệu chỉnh Salinity: ",data['Salinity'].var())  #Phương sai chưa hiệu chỉnh
print("Phương sai hiệu chỉnh Salinity: ",data['Salinity'].var(ddof=0))  #Phương sai hiệu chỉnh
print("Tứ phân vị:")
data['Salinity'].quantile([0.25, 0.5 , 0.75])
```

Mối tương quan

```python
import numpy
import matplotlib.pyplot as plt

x=data.iloc[:,0]
y=data.iloc[:,4]

plt.scatter(x,y)
plt.show
```

Phân tổ.

```python
data_agg =data.groupby('Year').agg(avg_salinity = ('Salinity','mean'),avg_temperature = ('Temperature','mean')).round()
data_agg.head()
data_agg.plot(kind='bar', title="Trung bình nhiệt độ và độ mặn theo từng năm").legend();
```
