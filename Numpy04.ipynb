Câu 1: Sử dụng thư viện Numpy, hãy tạo một vector a có 7 chiều (tương ứng 7 phần tử) với giá
trị các phần tử được sinh bằng hàm (kết quả xuất ra màn hình):
- np.arrange()
- np.linspace()
- np.logspace()
- np.random.rand()

Ghi chú: sau mỗi lần sinh ra giá trị cho vector, hãy cho biết số chiều của vector được sinh ra và
số chiều của vector chuyển vị của vector đó; cho biết kích thước của vector a và vector chuyển vị
của vector a.

```python
import numpy as np
a = np.arange(-3, 4, 1, dtype=int)
a1 = np.transpose(a)
b = np.linspace(0, 6, 7)
b1 = np.transpose(b)
c = np.logspace(1.0, 7.0, num=7, base = 2)
c1 = np.transpose(c)
d = np.random.rand(1,7)
d1 = np.transpose(d)
print(a)
print(a.ndim, a1.ndim, a.shape, a1.shape)
print(b)
print(b.ndim, b1.ndim, b.shape, b1.shape)
print(c)
print(c.ndim, c1.ndim, c.shape, c1.shape)
print(d)
print(d.ndim, d1.ndim, d.shape, d1.shape)
```

Câu 2: Sử dụng thư viện Numpy, hãy tạo ra ma trận A10x7 với giá trị các phần tử của ma trận A
được sinh bởi:
- np.arrange()
- np.linspace()
- np.logspace()
- np.random.rand()
Sau mỗi lần ma trận được tạo ra, hãy xuất kết quả ra màn hình và cho biết:
- Số chiều và kích thước của ma trận A và ma trận AT
.

```python
import numpy as np
a = np.arange(0, 70, 1)
a.shape = (10,7)
a1 = np.transpose(a)
b = np.linspace(0, 69, 70)
b.shape = (10,7)
b1 = np.transpose(b)
c = np.logspace(0, 10, num=70, base = 2)
c.shape = (10,7)
c1 = np.transpose(c)
d = np.random.rand(1, 70)
d.shape = (10,7)
d1 = np.transpose(d)
print(a)
print(a.ndim, a1.ndim, a.shape, a1.shape)
print(b)
print(b.ndim, b1.ndim, b.shape, b1.shape)
print(c)
print(c.ndim, c1.ndim, c.shape, c1.shape)
print(d)
print(d.ndim, d1.ndim, d.shape, d1.shape)
```

Câu 3: Sử dụng thư viện Numpy, hãy:
- Chuyển ma trận A ở câu 2 thành vector. (Gợi ý: sử dụng hàm flatten())
- Chuyển vector vừa được tạo ra ở câu trên thành ma trận. (Gợi ý: sử dụng hàm reshape())

```python
import numpy as np
a = np.arange(0, 70, 1)
a.shape = (10,7)
a1 = a.flatten()
a2 = a1.reshape(10,7)
b = np.linspace(0, 69, 70)
b.shape = (10,7)
b1 = b.flatten()
b2 = b1.reshape(10,7)
c = np.logspace(0, 10, num=70, base = 2)
c.shape = (10,7)
c1 = c.flatten()
c2 = c1.reshape(10,7)
d = np.random.rand(1, 70)
d.shape = (10,7)
d1 = d.flatten()
d2 = d1.reshape(10,7)
print(a)
print(a1)
print(a2)
print(b)
print(b1)
print(b2)
print(c)
print(c1)
print(c2)
print(d)
print(d1)
print(d2)
```

Câu 4: Sử dụng thư viện Numpy, hãy:
- Tạo ra 1 ma trận A10x15 với các giá trị được sinh ra bởi hàm logspace();
- Hãy trích rút dòng thứ 3 của ma trận A (tương ứng với 1 vector hàng);
- Hãy trích rút ra cột thứ 5 của ma trận A (tương ứng với 1 vector cột).
